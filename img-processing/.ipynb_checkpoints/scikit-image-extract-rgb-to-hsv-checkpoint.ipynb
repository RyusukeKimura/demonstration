```python
# import libraries
import matplotlib.pyplot as plt
from skimage.color import rgb2hsv
from skimage import data, exposure, io, filters
%matplotlib inline
```

```python
# read data
image = data.astronaut()
```

```python
# show original image
io.imshow(image)
io.show()
```

```python
# rgb to hsv
hsv = rgb2hsv(image)
```

```python
# show hsv
io.imshow(hsv)
io.show()
```

```python
# decompose
hue = hsv[:, :, 0] # hue(色相)
sat = hsv[:, :, 1] # saturation(彩度)
val = hsv[:, :, 2] # value(明度)
```

```python
# show val image
io.imshow(val)
io.show()
```
