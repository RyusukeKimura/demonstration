```python
# import libraries
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure, io, filters
from skimage.transform import rescale, resize, downscale_local_mean
%matplotlib inline
```

```python
# read data
coffee = data.coffee()
cat = data.chelsea()
```

```python
# check image shape
print(coffee.shape)
print(cat.shape)
```

```python
# resize image
coffee = resize(coffee, (cat.shape[0], cat.shape[1], cat.shape[2]), anti_aliasing=True)
```

```python
# show images
## coffee
io.imshow(coffee)
io.show()

## cat
io.imshow(cat)
io.show()
```

```python
# convert to gray scale
from skimage.color import rgb2gray
coffee_grayscale = rgb2gray(coffee)
cat_grayscale = rgb2gray(cat)
```

```python
# show images
## coffee
io.imshow(coffee_grayscale)
io.show()

## cat
io.imshow(cat_grayscale)
io.show()
```

```python
# substract image
substraction = coffee_grayscale - cat_grayscale
```

```python
# show substracted image
io.imshow(substraction)
io.show()
```
