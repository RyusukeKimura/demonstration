```python
# import libraries
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, exposure, io, filters
%matplotlib inline
```

```python
# read data
image = data.astronaut()
```

```python
# show original image
io.imshow(image)
io.show()
```

```python
# extract each color
red = image[: , : , 0]
green = image[: , : , 1]
blue = image[: , : , 2]
```

```python
# show each color image
io.imshow(green)
io.show()
```

```python
# make image green!!
green = image.copy()
green[: , :, 0] = 0
green[: , :, 2] = 0
```

```python
# show green image
io.imshow(green)
io.show()
```
