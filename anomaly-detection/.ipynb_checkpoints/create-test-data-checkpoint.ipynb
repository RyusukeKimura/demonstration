```python
# import libralies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# omajinai
pd.set_option('display.max_rows', 500)
```

```python
# set params
## number of records
n = 1000

## 1D
mu1 = 0
sigma1 = 30

## 2D
mu2 = [0, 0]
sigma2 = [[30, 20], 
                 [20, 50]]
```

```python
# create values
## 1D
values1 =  np.random.normal(mu1, sigma1, n)
plt.hist(values1, bins=50)
plt.show()

## 2D
values2 = np.random.multivariate_normal(mu2, sigma2, n)
sns.jointplot(values2[:,0], values2[:,1])
plt.show()
```

```python
# write test data
np.savetxt('./data/normal_dist_1d.csv', values1, delimiter=',', header='x', comments='')
np.savetxt('./data/normal_dist_2d.csv', values2, delimiter=',', header='x, y', comments='')
```
